```python
# Helper function to extract ONLY parsed_value (article keywords extracted) and their relevance in text

def extract_keywords_relevance(body):
    my_extractions_list = body['extractions']

    keyword_relevance_json = {} # key = keyword, value = it`s relevance score inside the data text

    for extract in my_extractions_list:
        keyword_relevance_json[extract['parsed_value']] = extract['relevance']

    return keyword_relevance_json

# call fcn
# keyword_relevance_mapping = extract_keywords_relevance(body)
```

```python
# Helper function to extract ALL keywords for ALL articles

keyword_json = {} # json, key = link of the article, value = json, key = keyword, value = its relevance in the data text
link_processingtime_mapping = {}  # json, key = link of article, value = processing time for keyword - relevance extraction task
index = 1 # index used for counting the article range from db

starttime = datetime.now() # START processing time for extracting keywords from ALL articles
print('Script #2 starts at: ', starttime)

# go through ALL the articles found in db
for link, content in db_without_bl.items(): # where key = article_url and value = text content
    if index >= 1 and index <= 7626: # Could not run with len(db), because free req limit is 1K req / month / account ! Always same order in dict !
        print('working on article: ', link)
        print(index)

        before_processing_datetime = datetime.now() # START processing time for an article

        # getting the json response of the MonkeyLearn API for KeywordExtractor request
        body = extract_keywords_from_text(content)

        # getting the article keywords extracted and their relevance in text
        keyword_json[link] = extract_keywords_relevance(body) # where key = article_url and value = 10 keywords extracted from text content

        after_processing_datetime = datetime.now() # STOP processing time for an article
        diff_datetime = after_processing_datetime - before_processing_datetime
        link_processingtime_mapping[link] = diff_datetime

    index = index + 1 # increment for the next article text content

endtime = datetime.now() # STOP processing time for extracting keywords from ALL articles
print('Script #2 ends at: ', endtime)
```
